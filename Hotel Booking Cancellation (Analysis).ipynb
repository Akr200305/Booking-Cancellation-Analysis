# Data Analysis & Visualizations

### 1. What amount of reservations are canceled and not-canceled

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
```

#### 37% of total reservations are cancelled

### Lets visualize this part of the data

```python
plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels, _ = ax1.get_legend_handles_labels()

ax1.legend(legend_labels, labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))

plt.title('Reservation Status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

#### Awesome!!
### Now we want to know the percent of hotels in each of the Resort Hotel and City Hotel are getting cancelled.

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

#### Percentage of hotel cancellation in City Hotel are much larger than that of Resort Hotel.
### Now we will check if the prices of booking in these has affected these bizarre cancellation rates.

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')

plt.xlabel('Date or Time Period')
plt.ylabel('Average Daily Rate (ADR)')

# Add a legend
plt.legend(fontsize = 20)

plt.show()
```

#### So, I had this hypothesis in mind that the prices of city hotel are less than resort hotel. And that's why people tend to cancel the booking more in city hotel. Because it would cost less cancellation charges.
### This proves the hypothes is correct that on some or most days, prices of city hotel is less than resort hotels.

## Now, which months of the year has most and least reservations and cancellations.

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')

legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1, 1))
plt.title('Reservation Status per Month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

### From the above countplot we can analyse few points -
#### - Maximum cancellations in January
#### - Minimum cancellations in August and also the Maximum reservations

### Hypothesis : 
#### Since we got huge cancellations in January and not so much in August. So, maybe this is beacuse the prices were high in Jan but not so much in August.

```python
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize = 20)

df_canceled = df[df['is_canceled'] == 1]
df_monthly_sum = df_canceled.groupby('month')['adr'].sum().reset_index()

sns.barplot(x = 'month', y= 'adr', data = df_monthly_sum)
plt.show()
```

### We can see that Adr in Jan is the highest and lowest in August. Accoording to the hypothesis we predicted this.

#### Because people generally think that we can get some cheaper hotels and even got that and then cancel this one.

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with most cancellations')
plt.pie(top_10_country, autopct = '%0.2f', labels = top_10_country.index)
plt.show()
```

## Now, we want to calculate whose Adr is more -
### -- the one with the cancelled reservations 
####   or
### -- the one with the non-cancelled reservations

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = cancelled_data = df[df['is_canceled'] == 0]

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 15)
plt.show()
```

## It's quite clear that as the Average Daily Rate increases, cancellation rate increases.
